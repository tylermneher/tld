```python
import pandas as pd
import requests
```

```python
url = "https://tld-list.com/df/tld-list-details.json"
```

```python
response = requests.get(url)
json_response = response.json()

df = pd.DataFrame.from_dict(json_response)
tld_df = df.transpose()

tld_df
```

```python
tld_df.to_csv('tld_df.csv', index=True, index_label='tld')
!cat tld_df.csv
```
